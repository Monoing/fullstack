# **Group by Aggregate**

```python
df.head()
```

```python
##########
df["price"].value_counts()
```

```python
df.groupby(["price"])[["median_income"]].agg(["min", "mean", "median", "max", "std"])
```

```python
df["median_income"].describe()
```

```python
### กด ctrl + D เพื่อเลือแก้คำซ้ำพร้อมกัน
def transform_income(_income):
    if _income <= 2.56:
        return "poor"
    elif _income >= 4.76:
        return "rich"
    else:
        return "middle"
```

```python
df["income"] = df["median_income"].apply(transform_income)
```

```python
df[["income","median_income"]].sample(5)
```

```python
df_group_price_income = df.groupby(["price", "income"])[["median_income"]]\
    .agg(["count","min", "mean", "median", "max", "std"])
```

```python
df_group_price_income
```

```python
df_group_price_income.reset_index().style.background_gradient()
```

#**Pivotable**

```python
df_pivot_price_income = pd.pivot_table(data = df, 
               values="median_income", 
               index="price", 
               columns="income", 
               aggfunc='mean', 
               fill_value=None)
```

```python
df_pivot_price_income
```

```python
df_pivot_price_income.style.background_gradient()
```
